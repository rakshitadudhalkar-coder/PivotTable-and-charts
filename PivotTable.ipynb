EASY LEVEL
#1.What is the purpose of a PivotTable in Excel?
Ans -

A PivotTable is used to summarize, analyze, and organize large datasets quickly.
It helps in:
Calculating totals, averages, counts
Comparing data (e.g., sales by region)
Creating reports without changing original data.

#2.Create a PivotTable to show Total Sales by Region.
Ans -

 1.Select the dataset.

 2.Go to INSERT - PIVOTTABLE

 3.Choose new worksheet - click OK.

 4.Drag Region - Rows, Sales - Values(Sum of Sales).

 5.We'll get output as Total sales for East,West,North and South.

#4.Explain the difference between Sorting and Filtering with an example from this dataset.

Ans -

Sorting

Arranges data (Ascending/Descending)

Example: Sort Sales from highest to lowest

Filtering

Shows only specific data

Example: Show only Sales > 300

# Example from dataset:

Sorting: Arrange regions by highest sales

Filtering: Show only “North” region sales
